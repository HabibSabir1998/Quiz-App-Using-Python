# QUIZ APP

```python
questions = [
    {
        "qno":1,
        "question":"How to get input from user?",
        "options":["scanf()","parseint()","get()","input()"],
        "answer": "input()"
        
    },
    
    {
        "qno":2,
        "question":"Index Start from?",
        "options":["0","1","-1","2"],
        "answer": "0"
        
    },
    
    {
        "qno":3,
        "question":"What is the output of print 0.1 + 0.2 == 0.3?",
        "options":["True","False","Machine Dependent","Error"],
        "answer": "False"
        
    },
    
    {
        "qno":4,
        "question":" Tuple Define By ____________ Brakets",
        "options":["[]","{}","()","<>"],
        "answer": "()"
        
    },
    
    {
        "qno":5,
        "question":"""What will be the output of the following Python statement?
                        >>> print('new' 'line')""",
        "options":[" Error","Output equivalent to print ‘new\nline’","newline","new line"],
        "answer": "newline"
        
    },
    
]
score = 0
for question in questions:
    data = str(question["qno"]) + ": " + question["question"] + "\n Please write answer in input field\n\n" + "a) " + question["options"][0] + "\n"+ "b) " + question["options"][1] + "\n"+ "c) " + question["options"][2] + "\n"+ "d) " + question["options"][3] + "\n\n"
    answer = input(data)
    print("\n")
    if answer == question["answer"]:
        score +=10
per = score*100/50
print("your score is: " ,str(score),"\n") 
print("you Got ",str(per),"%\n")

if (per <= 100 and per >= 70): 
    print("Yahooooo!!!!\n")
    print("A Grade")
elif (per <= 69 and per >= 60):
    print("Hurrraahhh!!!!\n")
    print("B Grade")
elif(per <=59 and per >=50):
    print("Keep it up\n")
    print("C Grade")
else:
    print("Better luck next time\n")
    print("Average")
```
